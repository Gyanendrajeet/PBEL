# Confusion Matrix

It is a Small table That actual labels with predicted labels

1. True Positive: predicted Positive, actually positive
2. True Negative: Predicted negative, actually Negative
3. False Positive: Predicted Positive actually negative
4. false Negative: predicts negative actually positive

# Example

Confusion Matrix is used to evaluate the performance of a classification model. It compares the **Actual Values** with the **Predicted Values**.

|                | Predicted Positive | Predicted Negative |
|----------------|-------------------|-------------------|
| **Actual Positive** | TP (True Positive) | FN (False Negative) |
| **Actual Negative** | FP (False Positive) | TN (True Negative) |

## 1. True Positive (TP) ✅
- Actual = Positive
- Predicted = Positive
- **Example:** Spam → Predicted as Spam

## 2. True Negative (TN) ✅
- Actual = Negative
- Predicted = Negative
- **Example:** Not Spam → Predicted as Not Spam

## 3. False Positive (FP) ❌
- Actual = Negative
- Predicted = Positive
- **Example:** Not Spam → Predicted as Spam

## 4. False Negative (FN) ❌
- Actual = Positive
- Predicted = Negative
- **Example:** Spam → Predicted as Not Spam

## Why do we use Confusion Matrix?
- To identify correct and incorrect predictions.
- To know the types of errors made by the model.
- To calculate Accuracy, Precision, Recall, and F1-Score.
