```python
# data type 
# list 
my_list = [1, 2, 3, 4, ]

# methods 
print(my_list)
my_list.append(6) # add an element to the end of the list
print(my_list)


my_list.insert(2, 10) # insert 10 at index 2 
print(my_list)

my_list.remove(3) # the first occurence of 3 
print(my_list)


my_list.pop() # Remove and return the last element 
print(my_list)


# properties
# ordered and mutable
my_list[2]= 20 # modifies the elemnt at index 2 
print(my_list)


#  hetrogeneous & duplicates allowed

zen=[1,2,3,4,5,6, "ai", True, False ]
print (zen) # output: [1,2,3,4,5,6,"ai", True, False]
```
