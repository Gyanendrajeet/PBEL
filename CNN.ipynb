```python
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
import numpy as np
```

```python
(X_train,Y_train),(X_test,Y_test)= keras.datasets.cifar10.load_data()
```

```python
Y_train = Y_train.reshape(-1,)
Y_train[:5]
```

```python
Y_test = Y_test.reshape(-1,)
```

```python
classes = ["Airplane","Automobile", "Bird", "Cat", "Deer", "Frog", "Horse", "Ship", "Truck"]
```

```python
def plot_sample(X,Y,index):
  plt.figure(figsize=(15,2))
  plt.imshow(X[index])
  plt.xlabel(classes[Y[index]])
```

```python
plot_sample(X_train,Y_train,0)
```

```python
plot_sample(X_train,Y_train,-1)
```

```python
X_train_Scaled = X_train/255
X_test_Scaled = X_test/255
```

ANN

```python
ann = keras.models.Sequential([
    keras.layers.Flatten(input_shape=(32,32,3)),
    keras.layers.Dense(3000,activation='relu'),
    keras.layers.Dense(1000,activation='relu'),
    keras.layers.Dense(10,activation='softmax')

])

ann.compile(
    optimizer='SGD',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']

)
ann.fit(X_train_Scaled,Y_train,epochs=10)
```

CNN

```python
cnn=keras.models.Sequential([
    keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',
                        input_shape=(32,32,3)),
    keras.layers.MaxPooling2D(2,2),

    keras.layers.Flatten(),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(10,activation='softmax')
])
cnn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

cnn.fit(X_train_Scaled,Y_train,epochs=10)
```

```python
cnn.evaluate(X_test_Scaled,Y_test)
```

```python
cnn1=keras.models.Sequential([
    keras.layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3)),
    keras.layers.MaxPooling2D((2,2)),

    keras.layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'),
    keras.layers.MaxPooling2D((2,2)),


    keras.layers.Flatten(),
    keras.layers.Dense(64,activation='relu'),
    keras.layers.Dense(10,activation='softmax')

])

cnn1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

cnn1.fit(X_train_Scaled,Y_train,epochs=10)
```

```python
cnn1.evaluate(X_test_Scaled,Y_test)
```

```python
y_pred = cnn.predict(X_test)
y_pred[:5]
```

```python
y_classes = [np.argmax(element) for element in y_pred]
y_classes[:5]
```
