#### What is Gini Index ?

- Gini Index measures "How impure or mixed a dataset is"
- More mixed data ----->> Higher Gini
- More pure data ------>> Lower Gini

- Decision tree always try to Reduce GINI 

- Entropy ask "How much information is missing?"
- Gini aks "How likely am i to classify a incorrectly?"
- Both reaches almost the same decision .
- But Gini requires fewer calcu.
- Therefore, Gini is faster.

#### Basic Libreries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

```

```python
from sklearn.model_selection import train_test_split # sklearn = library, model_selection = module, 
# suffle + train and test split 
from sklearn.datasets import load_breast_cancer 
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score,recall_score,f1_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

#### Load Dataset 

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df ["Target"] = cancer.target
```

```python
df.head(10)
```

#### Explore data 

```python
print("shape:", df.shape)
```

```python
print(df.info())
```

```python
print(df['Target'].value_counts())
```

#### Seperate features and target 

```python
x= df.drop('Target', axis=1)
y = df["Target"]
```

#### Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print("training shape:", x_train.shape)
```

```python
print("Testing Shape:", x_test.shape)
```

#### Decision Tree Model create

```python
model = DecisionTreeClassifier(
    criterion = "gini",
    random_state=42
)
```

#### Train model

```python
model.fit(x_train, y_train)
```

#### Predict 

```python
y_pred = model.predict(x_test)
```

```python
comparision = pd.DataFrame({
    'Actual':y_test.values,
    "predict": y_pred
})
```

```python
comparision.head(10)
```

#### Accuracy 

```python
accuracy = accuracy_score(y_test,y_pred)
print("Acc:",round(accuracy,2))
```

#### Precision, Recall,f1

```python
Precision = precision_score(y_test, y_pred)
print ("Precision:",round(Precision,2))

Recall = recall_score(y_test,y_pred)
print("Recall:",round(Recall,2))


f1 = f1_score(y_test,y_pred)
print(f1)
```

#### Confusion matrix

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(cm,annot=True, cmap="Blues", fmt= 'd',xticklabels=['Maligant',"benign"], yticklabels=['maligant', "benign"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Decision Tree confusion matrix")
plt.show()
```

#### Classification Report

```python
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(20,10))

plot_tree(
    model,
    feature_names=x.columns,
    class_names=['Malignant', 'Benign'],
    fontsize=9,
    rounded=True,
    filled=True
)

plt.show()
```
