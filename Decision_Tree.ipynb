# Understand structure of our decision tree

- Real Tree: it has -> Roots -> trunk -> branches -> leaves
- Decision tree: Roots -> Question -> More questions -> Final answer
- A Decision tree grows exactly like a natural tree. Only difference is that instead of growing leaves it grows decision.


#### Parent Node:
 - Parent Nodde creates on or more child nodes
 - Salary(parent node) -> yes or no -> child node

#### Child Node:
 - Every node from another node is called a child node.

Root node -> Internal Node -> Braanch -> child Node -> Leaf Node
 - This entire stucture is called a Decision Tree!
 

#### Pure cs Impure Nodes

- Basket A: u hv apple, apple,apple,apple
- Pure

- Basket B: apple, banana, apple, orange, banana, apple
- Impure

- Tree loves pure nodes
- Tree hate impure nodes
- The goal is simple convert impure -> less impure -> almost pure -> cmpletely pure

- [Decision Tree grow by increasing purity].
